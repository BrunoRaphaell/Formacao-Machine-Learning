# 02. K-Means

K-Means é um algoritmo de clusterização (ou agrupamento). É um algoritmo de aprendizado não supervisionado, que avalia e clusteriza os dados de acordo com suas caracteristicas.

O K-means não funciona bem quando os nossos clusters não têm simetria radial, ou seja, eles não são próximos de uma bolinha, quando não dá para nós separarmos ele com uma bolinha


Por que usar dados normalizados? Quando temos dados com distâncias muito grandes, como é o caso desse dataset onde temos que o máximo do Álcool é 14.83 e do Proline é de 1680, ou seja, tamanhos extremamente diferentes, é interessante normalizar os dados para que eles tenham uma variação de 0 a 1 para que seja melhor de analisar e fazer as clusters.

# 03. DBSCAN

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a popular unsupervised learning method utilized in model building and machine learning algorithms.

- [Artigo](https://ml-explained.com/blog/dbscan-explained)

![](https://miro.medium.com/max/724/0*PQTm0HWuxQHuxHyl.gif)

1. Vantagens:

- Robusto a outliers;
- Pode detectar qualquer tipo de forma; 
- Custo computacional não é tão alto;
- Segundo algoritmo de clusterização mais utilizado;

2. Desvantagens:

- Não funciona bem com dados de alta dimensionalidade
- Não consegue identificar clusters de várias densidades, uma vez que os parâmetros *eps* e *min_samples* são fixos.
- A paratetrização de *eps* e *min_samples* nem sempre é uma tarefa fácil.

```python
from sklearn.cluster import DBSCAN

"""
É muito interessante que diferentemente do KMeans nós não passamos o número de cluster, 
o algoritmo automaticamente nos definiu quantos grupos ele acha que são necessários, 
respeitando essas duas regras que nós colocamos.

eps: Significa a distância máxima entre dois pontos para que eles possam ser considerados vizinhos ou pertencentes ao mesmo cluster
min_samples: Número mínimo de pontos para que eu posso formar um cluster.
metric: Métrica de distância, ou seja, uma forma de avaliar a distância entre os pontos.
"""

agrupador = DBSCAN(eps = 1.31, min_samples = 15, metric='manhattan')
agrupador.fit(df)
```

```python
"""
-1 significa que a amostra foi classificada como ruído.
porque o DBSCAN é feio para aplicações com ruído, ele considera que qualquer ponto que não esteja dentro dessa distância de 1.31, que é a máxima dos grupos, ele considera que isso é um ruído
"""

agrupador.labels_
```

```python
# Valores únicos do agrupador. 

np.unique(agrupador.labels_)
```

Métricas [aceitáveis](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise_distances.html#sklearn.metrics.pairwise_distances) para o parâmetro *metric*.

![](https://iq.opengenus.org/content/images/2018/12/distance.jpg)

O *eps* é sempre calculado em cima de uma métrica de distância. 

O KMeans trabalha com a distância euclidiana.

Vamos analisar agora como a clusterização, o agrupamento realizado pelo DBSCAN, é sensível aos parâmetros do próprio modelo.

```python
# Com a métrica manhattan

agrupador = DBSCAN(eps = 1.5, min_samples = 1, metric='manhattan')
agrupador.fit(df)
agrupador.labels_
```

```python
# Com a métrica euclidiana

agrupador = DBSCAN(eps = 0.5, min_samples = 15, metric='euclidean')
agrupador.fit(df)
agrupador.labels_
```

Existem outras métricas de distância, além das apresentadas aqui, como distância cosseno e outras. Para alterar no algoritmo, basta fazer como na aula, alterar o parâmetro metric. As métricas de distância permitidas são:

- Do scikit-learn: [‘cityblock’, ‘cosine’, ‘euclidean’, ‘l1’, ‘l2’, ‘manhattan’]
- Do scipy.spatial.distance: [‘braycurtis’, ‘canberra’, ‘chebyshev’, ‘correlation’, ‘dice’, ‘hamming’, ‘jaccard’, ‘kulsinski’, ‘mahalanobis’, ‘minkowski’, ‘rogerstanimoto’, ‘russellrao’, ‘seuclidean’, ‘sokalmichener’, ‘sokalsneath’, ‘sqeuclidean’, ‘yule’]

Basicamente, o que nós pudemos perceber através disso? O “eps” e o “min_samples” variam de maneira inversamente proporcional, ou seja, o efeito que eu causo aumentando o “eps” é o mesmo efeito que eu causo diminuindo o “min_samples”. O efeito que eu causo aumentando o “min_samples” é o mesmo efeito que eu causo diminuindo o “eps”.

A métrica de distância influencia bastante. Por quê? Porque ela influencia na forma como eu calculo o “eps”.



# 04. Mean Shift

- [DOCUMENTAÇÃO](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.MeanShift.html)

![](https://spin.atomicobject.com/wp-content/uploads/ms_2d_bw_.8.gif)

![](https://spin.atomicobject.com/wp-content/uploads/ms_2d_bw_2.gif)

Depois de ter que parametrizar tanta coisa com o DBSCAN, uma coisa que nós deveríamos nos perguntar é se não existe nenhum método de clusterização do qual nós pudéssemos utilizar sem termos que parametrizarmos tanta coisa, ou sem ter que parametrizarmos nada.

O Mean-Shift é um algoritmo de clusterização que diferentemente do KMeans, ele não requer especificação do número de cluster e diferentemente do DBSCAN você não precisa, por exemplo, parametrizar coisas como “eps” - no caso era o “epsilon” do DBSCAN - e o número mínimo de amostras por cluster, mas semelhantemente ao DBSCAN, ele procura por regiões de alta densidade e afirma que nessas regiões de alta densidade existem clusters.

Objetivo:
- Agrupar os pontos de acordo com as regiões de alta densidade

Por que nós deveríamos utilizá-lo?
- Essa é uma pergunta muito boa! Primeiramente, o K-means não costuma apresentar um bom resultado quando há ruído e grupos de simetria não radial, como nós vimos em aulas anteriores. Ele tem menos parâmetros para ser ajustado que o DBSCAN. O DBSCAN tem o “eps”, o “min_samples” e a métrica do cálculo de distância

Vantagens:
- Robusto a outliers;
- Serve para vários tipos de aplicações;
- Pode detectar qualquer tipo de forma;
- Tem apenas um parâmetro para ser ajustado, ou seja, o “bandwidth”.

Desvantagens: 
- Não funciona bem com dados de alta dimensionalidade;
- A parametrização da largura de banda não é tão simples;
- É computacional caro, porque todo ponto no início é interpreatado como cluster;




```python
from sklearn.cluster import MeanShift

agrupador = MeanShift()
agrupador.fit(df)
agrupador.labels_
```

```python
"""
O problema é que às vezes sem passar parâmetro, ele não funciona muito bem. 
Seria interessante nós sabermos se nós poderíamos passar algum parâmetro para ele, 
para ele funcionar bem.

Para nós fazermos o teste de como que a nossa clusterização varia de acordo com a 
largura de banda, nós vamos utilizar uma função pronta do scikit-learn, chamada: “estimate_bandwidth” 

O quantil define o ponto e qual é basicamente o tamanho da região de busca. 
Um quantil pequeno, uma busca pequena, grande número de clusters. 
Um quantil grande, busca grande, pequeno número de clusters, ou seja, poucos clusters
"""

from sklearn.cluster import estimate_bandwidth

BW = estimate_bandwidth(df, quantile=0.2)
agrupador = MeanShift(bandwidth=BW)
agrupador.fit(df)
agrupador.labels_
```

# 05. Coeficiente de silhueta para avaliação de clusterizações

 Depois de realizar todos esses tipos de clusterização, K-means, DBSCAN e MeanShift, você gostaria de saber se existe alguma métrica que indicasse que a clusterização que você fez está boa

Ele vai avaliar uma coisa chamada coesão. O quanto eu estou próximo aos elementos do meu grupo. Quanto maior a coesão, melhor o coeficiente de silhueta e de similaridade, o quanto eu estou afastado dos elementos do outro bairro. Quanto mais afastado eu estiver deles, melhor.

Objetivo:
- Avaliar os pontos em uma escalada de -1 a 1 em relação: 
    - Ao grupo (coesão)
    - Aos demais pontos (dissimilaridade)
- Quanto maior o valor da silhueta dos pontos, melhor.

```python
agrupador = KMeans(n_clusters=5)
agrupador.fit(df)
labels = agrupador.labels_
labels
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x = df['Intensidade_de_cor'], 
                           y = df['Alcool'], 
                           z = df['Proline'], 
                           mode = 'markers', 
                           marker = dict(color = labels.astype(float)), 
                           text = labels))
fig.show()
```

```python
from sklearn.metrics import silhouette_score

faixa_n_clusters = [i for i in range(2, 10)]
faixa_n_clusters
```

```python
from sklearn.cluster import KMeans

valores_silhueta = []

for k in faixa_n_clusters:
    agrupador = KMeans(n_clusters=k)
    labels = agrupador.fit_predict(df)
    valores_silhueta.append(silhouette_score(df, labels))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=faixa_n_clusters, y=valores_silhueta))
fig.update_layout(title_text='Valores de Silhueta Médio', xaxis_title='Número de clusters', yaxis_title='Valor de Silhueta Médio')
fig.show()
```

```python
"""
Solução apresentada pelo aluno João Francisco Baiochi:
https://cursos.alura.com.br/forum/topico-metodo-alternativo-de-plotar-207018 
"""

silhouette_values = []
for quantile in np.linspace(0.02, 0.5, 30):
    bw = estimate_bandwidth(X=df, quantile=quantile)
    agg = MeanShift(bandwidth=bw).fit(df)
    try:
        score = silhouette_score(df, agg.predict(df))
        silhouette_values.append([score, quantile])
    except ValueError:
        silhouette_values.append([0, quantile])

silhouette_values_df = pd.DataFrame(silhouette_values, columns=['score','quantile'])

fig = go.Figure()
fig.add_trace(go.Scatter(x = silhouette_values_df['quantile'],
                         y = silhouette_values_df['score'])
)
fig.update_layout(dict(title = 'Silhouette values',
                xaxis_title = 'Quantile',
                           yaxis_title='Silhouette score')
)
fig.show()
```

Então, o que nós podemos tirar de lição com isso? Sempre que nós formos aplicar alguma coisa relacionada clusterização, nós não devemos utilizar apenas uma técnica, mas várias e depois avaliarmos qual delas produziu o melhor resultado segundo algum tipo de análise, que no nosso caso foi o coeficiente de silhueta. Uma vez analisado isso, você define a melhor técnica e utiliza essa para agrupar os seus dados.

```python
# Definindo os melhores valores para eps e min_samples

#20 valores para cada parâmetro.
faixa_min_samples = [i for i in range(1,100,5)]
faixa_eps = [i/10 for i in range(1,100,1)]
```

Outro ponto interessante ao qual devemos atentar é que a silhueta não pode ser calculada quando o número de clústeres é igual a 1 ou igual ao número de dados, no caso 178. No primeiro caso não existem vizinhos para comparar o grau de pertença ao grupo. No segundo caso cada cluster só possui um ponto, então, não é possível calcular a semelhança do mesmo com ele mesmo e com os demais. Para evitar esses problemas, então, colocaremos um if, ou "se”, que basicamente avaliará se o número de labels é diferente de 1 e de 178. Se sim, ele calculará a silhueta. Para contar o número de labels, utilizaremos uma função chamada counter da biblioteca collections, que retorna um array onde a primeira posição é o valor e a segunda é o número de ocorrências ou o número de repetições.

```python
from collections import Counter
```

```python
valores_silhueta = []
eps_plot = []
min_samples_plot = []
for min_samples in faixa_min_samples:
  for eps in faixa_eps:
    labels = 0
    agrupador = DBSCAN(eps = eps, min_samples = min_samples, metric = 'manhattan')
    agrupador.fit_predict(df)
    labels = agrupador.labels_
    if(Counter(labels)[0] < len(df) and Counter(labels)[0] > 1): #porque não é possível calcular
    # silhueta para apenas  um cluster ou para número de clústeres igual
    # ao número de dados 
      media_silhueta = silhouette_score(df, labels)
      valores_silhueta.append(media_silhueta)
      eps_plot.append(eps)
      min_samples_plot.append(min_samples)
```
